## Import

```python
import matplotlib
import matplotlib.pyplot as plt
```

### Representacion Grafica de los Datos
Si a la funcion de trazado se le da una matriz de datos usara como coordenads en el eje vertical, y utilizara el índice de cada punto de datos en el array como la coordenda horizontal

```python
plt.plot([1, 2, 5, 7, 7, 8, 3, 1])
plt.show()
```

Tambien se puede proporcionar dos matricez: Un para el eje horizontal y otra para el eje vetical

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

Puede modificarse las longitudes de los ejes para que la figura no se vea tan ajustada

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 4, 0, 10]) #[xmin, xmax, ymin, ymax]
plt.show()
```

 Se sigue el mismo procedimiento para pintar una funcion matematica 

```python
import numpy as np 
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

Tambier puede modificarse el estilo de la grafica para que contenga mas informacion

```python
plt.plot(x, y)
plt.title("Squere Funcion")
plt.xlabel("Mate")
plt.ylabel("y = x²")
plt.grid(True)
plt.show()
```

 Pueden superponerse graficas y cambiar el estilo de las lineas

```python
import numpy as np 
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', x, y2, 'g')
plt.show()
```

```python
# Separarlo en diferentes lineas las funciones
import numpy as np 
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.title("Square Funcion")
plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g')
plt.show()
```

```python
# Para poder diferenciar entre las funciones siempre es recomendable añladir una leyenda
import numpy as np 
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.title("Square Funcion")
plt.plot(x, y, 'b--', label = "x**2")
plt.plot(x, y2, 'g', label = "x + 1")
plt.legend(loc = "best") # la situa en la mejor localizacion
plt.show()
```

Tambien puede crearse dos graficas que no se superpongan, estas graficas se organizan en un grip y se denominan subplots

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1) #1 rows, 2 columns, 1st subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) #1 rows, 2 columns, 2st subplot
plt.plot(x, y2, 'g')

plt.show()
```

#### Scatter Plots

```python
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x, y)
plt.show()
```

```python
from numpy.random import rand 
x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c = 'blue')
plt.show()
```

#### Histográmas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211) # 2 rows, 1 column, 1st subplot
plt.hist(data, bins=10, rwidth=0.8)

plt.subplot(212) # 2 rows, 1 column, 2st subplot
plt.hist(data, bins=[1, 1.5, 2, 2.5, 3], rwidth=0.95)

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
